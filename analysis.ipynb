# Analysis of Survey Results

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

data = pd.read_csv("https://docs.google.com/spreadsheets/d/e/2PACX-1vShf7f554C9uZkpZHDAMXS9HYqnzVgeOpeihMdZc6F1lSHtMkWC1jw1fUF7_-6fQ3xDjyaqKOCLkNNO/pub?gid=888413642&single=true&output=csv")
data.columns = ["timestamp","raw","kinds","market","solve_open_ended"]
```

```python
data
```

```python
## Submissions

pd.to_datetime(data.timestamp)
plt.plot(data.timestamp,np.cumsum(np.ones(len(data))))
plt.xticks(rotation=270)
plt.title("Submissions by time")
plt.show()
```

```python
# Do you offer Raw Infrastructure

pie_data = [data[data["raw"] == "Yes"]["raw"].count(),data[data["raw"] == "No"]["raw"].count()]
pie_labels = ["Yes","No"]
plt.pie(pie_data,labels=pie_labels, autopct='%1.1f%%',shadow=True,explode = [0.1,0])
plt.title("Do offer raw infrastructure?")
plt.show()
```

```python
# Do you Market your services

pie_data = [data[data["market"] == "No"]["market"].count(),data[data["market"] != "No"]["market"].count()]
pie_labels = ["No","Other"]
plt.pie(pie_data,labels=pie_labels,autopct='%1.1f%%',shadow=True,explode = [0.1,0])
plt.title("Do you market your services")
plt.show()
```
