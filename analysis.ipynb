```python
## Submissions

subs = pd.to_datetime(data.timestamp)
subs.plot()
subs.show()
```

```python
# Do you offer Raw Infrastructure

pie_data = [data[data["market"] == "No"]["market"].count(),data[data["market"] != "No"]["market"].count()]
pie_labels = ["No","Other"]
plt.pie(pie_data,labels=pie_labels)

plt.show()
```
